```python
# from scipy.optimize import curve_fit
from io import StringIO
from sklearn import linear_model
```

```python
regr = linear_model.LinearRegression()
```

```python
regr.fit(np.array(LMC_F_x).reshape(-1, 1), np.array(LMC_F_y))
regr.fit(np.array(LMC_1_x).reshape(-1, 1), np.array(LMC_1_y))
regr.fit(np.array(SMC_F_x).reshape(-1, 1), np.array(SMC_F_y))
regr.fit(np.array(SMC_1_x).reshape(-1, 1), np.array(SMC_1_y))
```

```python
fig = figure(figsize=(6, 6))
ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(222)
ax3 = fig.add_subplot(223)
ax4 = fig.add_subplot(224)

ax1.scatter(np.array(LMC_F_x), np.array(LMC_F_y), marker='.')
ax2.scatter(np.array(LMC_1_x), np.array(LMC_1_y), marker='.')
ax3.scatter(np.array(SMC_F_x), np.array(SMC_F_y), marker='.')
ax4.scatter(np.array(SMC_1_x), np.array(SMC_1_y), marker='.')

ax1.plot(np.array(LMC_F_x), regr.predict(np.array(LMC_F_x).reshape(-1,1)), color='red', linewidth=1)
ax2.plot(np.array(LMC_1_x), regr.predict(np.array(LMC_1_x).reshape(-1,1)), color='red', linewidth=1)
ax3.plot(np.array(SMC_F_x), regr.predict(np.array(SMC_F_x).reshape(-1,1)), color='red', linewidth=1)
ax4.plot(np.array(SMC_1_x), regr.predict(np.array(SMC_1_x).reshape(-1,1)), color='red', linewidth=1)
```
